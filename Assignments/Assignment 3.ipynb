```python
# reimport the data with above 12 columns (ignore all other variables)
data3= data[['UNITID', 'CITY', 'INSTNM', 'ZIP', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10']]
data3
```

```python

data2 = data3[data3["ADM_RATE"]!=0]
data2.shape
```

```python
# Find out the 10 colleges that are hardest to get in (top 10)
data_ADM = data2.nsmallest(10, 'ADM_RATE')
data_ADM


    
```

```python
# 10. Find out the average in-state tuition of top 10
mean = data_ADM["TUITIONFEE_IN"].mean()
print(mean)

```

```python
# Find out the average earnings of graduates from the top 10
mean3 = data_ADM["TUITIONFEE_IN"].mean()
print(mean3)
```

```python
# Find out the 10 colleges that are easiest to get in (bottom 10)
data_easy = data2.nlargest(10, 'ADM_RATE')
data_easy
```

```python
# Find out the average in-state tuition of the bottom 10
mean_2= data_easy["TUITIONFEE_IN"].mean()
print(mean_2)
```

```python
# Find out the average earnings of graduates from the bottom 10
data_easy["MN_EARN_WNE_P10"] = data_easy["MN_EARN_WNE_P10"].fillna(0)
mean4 = data_easy["MN_EARN_WNE_P10"]
print(mean4)
```
